# **Bar Chart (vertica-default)**

We make use of bar chart when we want to compare the categories by showing the values

```python
import matplotlib.pyplot as plt

fruits = ["Apple", "Mango", "Banana", "Grapes"]

# Values for each category
sales = [120, 150, 90, 110]

plt.bar(fruits, sales, width=0.3)
plt.xlabel("Fruits")
plt.ylabel("Sales")
plt.title("Sales Report of Fruits")
plt.grid()

plt.show()
```

# **Bar chart - Horizontal**

```python
import matplotlib.pyplot as plt

fruits = ["Apple", "Mango", "Banana", "Grapes"]

# Values for each category
sales = [120, 150, 90, 110]

plt.barh(fruits, sales)
plt.xlabel("Fruits")
plt.ylabel("Sales")
plt.title("Sales Report of Fruits")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Categories
fruits = ["Apple", "Mango", "Banana", "Grapes"]

# Values for two stores
store1_sales = [120, 150, 90, 110]
store2_sales = [100, 140, 95, 130]

# Width of each bar
width = 0.3

# X positions for first set of bars
x1 = [i - width/2 for i in range(len(fruits))]
# X positions for second set of bars
x2 = [i + width/2 for i in range(len(fruits))]

# Plot bars
plt.bar(x1, store1_sales, width=width, label="Store 1", color='skyblue')
plt.bar(x2, store2_sales, width=width, label="Store 2", color='orange')

# Labels and title
plt.xlabel("Fruits")
plt.ylabel("Sales")
plt.title("Fruit Sales Comparison Between Two Stores")
plt.xticks(range(len(fruits)), fruits)  # Set category names on X-axis
plt.legend()
plt.grid(axis='y')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Categories
fruits = ["Apple", "Mango", "Banana", "Grapes"]

# Values for three stores
store1_sales = [120, 150, 90, 110]
store2_sales = [100, 140, 95, 130]
store3_sales = [110, 130, 85, 120]

# Width of each bar
width = 0.2  # Make smaller so 3 bars fit in one group

# X positions for bars
x1 = [i - width for i in range(len(fruits))]  # Store 1
x2 = [i for i in range(len(fruits))]          # Store 2
x3 = [i + width for i in range(len(fruits))]  # Store 3

# Plot bars
plt.bar(x1, store1_sales, width=width, label="Store 1", color='skyblue')
plt.bar(x2, store2_sales, width=width, label="Store 2", color='orange')
plt.bar(x3, store3_sales, width=width, label="Store 3", color='green')

# Labels and title
plt.xlabel("Fruits")
plt.ylabel("Sales")
plt.title("Fruit Sales Comparison Between 3 Stores")
plt.xticks(range(len(fruits)), fruits)  # Set category names
plt.legend()
plt.grid(axis='y')

plt.show()
```

# **Pie Chart**

We make use of pie chart when we want to show/ represent the data in percentage.

```python
import matplotlib.pyplot as plt

# Categories
categories = ["Cash", "Card", "UPI"]

# Values for each category
values = [150, 100, 50]

# Create pie chart
plt.pie(values, labels=categories)
plt.title("Payment Method Distribution")
plt.show()
```

```python
plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title("Payment Method Distribution")
plt.show()
```

```python
explode = [0, 0.1, 0]  # Second slice (Card) is slightly separated

plt.pie(values, labels=categories, autopct='%1.1f%%', explode=explode)
plt.title("Payment Method Distribution")s
plt.show()
```

# **Scatte plot**

Use scatter plots to find relationships, trends, and outliers between two numerical variables.

```python
import matplotlib.pyplot as plt

quantity = [1, 2, 3, 4, 5]
total_spent = [100, 200, 250, 400, 500]

plt.scatter(quantity, total_spent, color='blue')
plt.title("Quantity vs Total Spent")
plt.xlabel("Quantity")
plt.ylabel("Total Spent")
plt.grid(True)
plt.show()
```

# **Histogram**

Histogram that shows the distribution of single numerical values

To compare frequency of different ranges

```python

quantity = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 7]

# Create histogram
plt.hist(quantity, bins=6, color='skyblue', edgecolor='black')
plt.title("Distribution of Quantity Sold")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()
```
