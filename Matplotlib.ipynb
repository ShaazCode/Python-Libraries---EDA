Line Plot with labels and title

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [1500,2000,1200,1000,1800]

plt.plot(months, sales)

# labeling
plt.xlabel("Months")
plt.ylabel("Sales")

plt.title("Sales Report")
plt.show()
```

Customize graph
color, marker , style

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [1500,2000,1200,1000,1800]

# customization color, style, marker
plt.plot(months, sales, color='red', linestyle='--', marker='o')

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales Report")


plt.show()
```

**Multiple line plots in single graphs**



```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales_2023 = [100, 150, 200, 180, 220]
sales_2024 = [130, 160, 210, 190, 250]

plt.plot(months, sales_2023)
plt.plot(months, sales_2024)

plt.show()
```

plt.legend(): Display label for each line

plt.grid(True): Add grid lines

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales_2023 = [100, 150, 200, 180, 220]
sales_2024 = [130, 160, 210, 190, 250]

plt.plot(months, sales_2023, label='Sales 2023')
plt.plot(months, sales_2024, label= 'Sales 2024')

plt.legend()
plt.grid()

plt.show()
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
city1 = [25, 26, 28, 30, 29, 27, 26]
city2 = [22, 23, 25, 26, 28, 27, 24]

plt.plot(days, city1, marker='o', label="City A")
plt.plot(days, city2, marker='s', label="City B")

plt.title("Weekly Temperature Comparison")
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid(True)
plt.show()
```
