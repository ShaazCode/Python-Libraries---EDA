**Normalization**

Normalization is a technique to rescale your data so that all value should fall in same range like 0 to 1

1. Min Max Scaling

Formula:
  Xnormalizaed = X - Xmin / Xmax - Xmin


Original data: [10, 20, 30, 40, 50]

Min = 10, Max = 50

Normalized values:

Original	Normalized
10	(10−10)/(50−10) = 0
20	(20−10)/(50−10) = 10/40 = 0.25
30	(30−10)/(50−10) = 20/40 = 0.5
40	(40−10)/(50−10) = 30/40 = 0.75
50	(50−10)/(50−10) = 40/40 = 1

All values now lie between 0 and 1.

```python
data = [10, 20, 30, 40, 50]

X_min = min(data)
X_max = max(data)

normalized = [(x - X_min) / (X_max - X_min) for x in data]

print("Original Data:", data)
print("Normalized Data:", normalized)
```
